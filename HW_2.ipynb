```python
import pandas as pd
import numpy as np
import math
import scipy.stats as stats
from sklearn.preprocessing import MinMaxScaler
```

```python
path = "/content/drive/MyDrive/myFiles/wine.csv"

df = pd.read_csv(path)

float_cols = df.select_dtypes(include=['float64']).columns # get all float dtype columns
float_means = df[float_cols].mean()   # get mean of every feature
df = df.fillna(float_means) # fill every missing float value

my_cols = df.select_dtypes(include=['int64']).columns # get all int dtype columns
int_means = df[my_cols].mean()   # get mean of every feature
df = df.fillna(int_means.astype(int)) # fill every missing int value

c_size = df.columns.size    # get total number of features
check_arr = np.zeros(c_size) #init empty array


for index, column in enumerate (df.columns):  # getting max-min of every column

  ar = np.array(df[column])
  check_arr[index] = ar.max() - ar.min()


index = check_arr.argsort()[::-1][:3]     # sorting values (from max to low -->[::-1]) and getting array of their origin indexes.(we need only 3)

col_1 = df.iloc[:, index[0]] # get serieses of every feature we want to change
col_2 = df.iloc[:, index[1]]
col_3 = df.iloc[:, index[2]]

df
```

***Min Max Normalization for 'col_1' >>proline<<***

```python
scaler = MinMaxScaler()

col_1 = col_1.values.reshape(-1, 1)   # Creating 2D array (1 coulmn and -1 samples)
df.iloc[:, index[0]] = scaler.fit_transform(scaler.fit_transform(col_1))  # Data normalization --> (min_value = 0 && max_value = 1)

df
```

***Zscore standartization for 'col_2' >>magnesium<<***

```python
df.iloc[:,index[1]] = stats.zscore(col_2.values.reshape(-1,1))  # Zscore standartization

df
```

***Equal depth binning for col_3 >>alcalinity_of_ash<<***

```python
def eqDepthBin(arr, bins):
  arr = arr.sort_values(ascending=True) # Sort pandas column array
  splits = np.array_split(arr[::-1], bins)[::-1]  # Reverse array for getting a lonngest sub-array in the end, Create the splits also.

  for index, ar in enumerate(splits): # get origin sort of the values.
    splits[index] = ar[::-1]

  return splits
```

***Smoothing by bin boundaries***

```python
def boundSmooth(data):

  max = data.max()
  min = data.min()

  for i, val in enumerate(data):
    if val - min <= max-val:
      data.iat[i] = min
    else:
      data.iat[i] = max
```

```python
col_3 = col_3.sort_values(ascending=True) #sorting pandas object of some feature
bins = eqDepthBin(col_3, 3) #spliting column to 3 bins

for bin in bins:
  boundSmooth(bin) # smoothing of every bin we got
  col_3.update(bin) # updating object of our column with every bin after smooting

col_3.sort_index(inplace = True) # cancel sorting of our column`s object

df.update(col_3) # updating column in origin DataFrame
```
